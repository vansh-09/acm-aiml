```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('income.csv')
```

```python
df.head()
```

```python
plt.scatter(df['Age'], df['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_pred = km.fit_predict(df[['Age', 'Income($)']])
y_pred
```

```python
df['Cluster'] = y_pred
df.head()
```

```python
df1 = df[df['Cluster'] == 0]
df2 = df[df['Cluster'] == 1]
df3 = df[df['Cluster'] == 2]

plt.scatter(df1['Age'], df1['Income($)'], color='green')
plt.scatter(df2['Age'], df2['Income($)'], color='red')
plt.scatter(df3['Age'], df3['Income($)'], color='blue')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()
# Fit the scaler on both columns together
scaler.fit(df[['Age', 'Income($)']])

# Transform both columns together and then assign back
scaled_data = scaler.transform(df[['Age', 'Income($)']])
df['Age'] = scaled_data[:, 0]
df['Income($)'] = scaled_data[:, 1]


#error solved using co-pilot
```

```python
y_pred = km.fit_predict(df[['Age', 'Income($)']])
y_pred
```

```python
df['Cluster'] = y_pred
df.head()
```

```python
df1 = df[df['Cluster'] == 0]
df2 = df[df['Cluster'] == 1]
df3 = df[df['Cluster'] == 2]

plt.scatter(df1['Age'], df1['Income($)'], color='green')
plt.scatter(df2['Age'], df2['Income($)'], color='red')
plt.scatter(df3['Age'], df3['Income($)'], color='blue')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```
